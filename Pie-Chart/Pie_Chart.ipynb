## Pie Chart

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

```python
df_fuel_engine_type = pd.read_csv("C:/Users/sahar/Documents/Data/udemy/pie-chart/pie_chart_data.csv")
df_fuel_engine_type
```

```python
# Set the matplotlib color cycle using a seaborn palette
sns.set_palette('colorblind')
```

```python
plt.figure(figsize=(10,8))
plt.pie(df_fuel_engine_type['Number of Cars'] ,
         labels= df_fuel_engine_type['Engine Fuel Type'].values,
         autopct = '%.2f%%',
         textprops={'size': 'large',
         'fontweight': 'bold',
         'rotation':30,
         'color': 'w'})
plt.legend()
plt.title('Cars by Engine Fuel Type' , fontsize = 18 , fontweight = 'bold')
plt.show()
```
